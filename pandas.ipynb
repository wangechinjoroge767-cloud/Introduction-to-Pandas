# Introduction to pandas  
- Pandas is a Python library used for data analysis, data cleaning and data manipulation.  
- It helps us work with structured data especially data in table form, similar to Excel, SQL tables or CSV files.  

_Analogy_ Pandas is a Python Excel.   

In pandas you can:  
- Open data
- Filter through the data  
- Create new columns  
- Clean missing values  
- Sort data  
- Group data  
- Do calculations on the data.  

## Main Data Structures in Pandas  
### a). Series.  
A Series is like a single column in Excel.  

````python
import pandas as pd
students = {'Name':['Mark', 'Allan', 'Bill', 'Renee', 'Mash', 'Michael', 'Jules', 'Sam', 'Steven']}
student_names = pd.Series(students)

student_names.head()
````

### b). DataFrame.  
A DataFrame is like a full table with rows and columns.  
A DataFrame is a 2 dimensional data.  

````python
students = {'Name':['Mark', 'Allan', 'Bill', 'Renee', 'Mash', 'Michael', 'Jules', 'Sam', 'Steven'],
 'Grade':['A', 'B', 'B', 'A', 'C', 'C', 'B', 'A', 'B'],
 'Course':['DS', 'DE', 'DS', 'DS', 'DE', 'DE', 'DS', 'DS', 'DE']}  

student_grades = pd.DataFrame(students)  

student_grades
````

__Thresholds for missing values and handling missing values__  
_Thresholds_  
- Between 0% to 30/40% nulls  
    - Fill the nulls using mean, median(numerical features) subject to the distribution of the dataset.  
    - For the categorical features/columns fill with the mode.  

- More than 40%  
    - Drop the column  

- Less than 5% nulls/missing values  
    - Fill, drop the rows

## Summarising data
__Grouping Data__  
Grouping is used when we want to summarize data by category.   
```python  
df.groupby(categorical_column)[column you are working]  
```  

examples of operations you can do are:   
- counts .count()  
- averages .mean()  
- Middlemost .median()  
- totals    .sum()



````python
# Grouping the data furnishing; total of each furnishing category  
df.groupby("furnishing")["furnishing"].count()
````

````python
# Grouping the furnishing categories to get the average rent per category  
df.groupby("furnishing")["monthly_rent_kes"].mean()
````

````python
# Average rent per property type  
df.groupby("property_type")["monthly_rent_kes"].mean()
````
